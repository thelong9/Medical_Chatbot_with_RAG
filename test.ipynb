```python
import dotenv
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma


REVIEW_CHROMA_PATH = "chroma_data/"
dotenv.load_dotenv()

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

reviews_vector_db = Chroma(
    persist_directory=REVIEW_CHROMA_PATH,
    embedding_function=embeddings,
)

question = """Has anyone felt frustrated about the service of the hospital?"""
k=10
relevant_docs = reviews_vector_db.similarity_search(question, k)

for i in range(k): 
    print(relevant_docs[i].page_content)
    print("/n")
```
